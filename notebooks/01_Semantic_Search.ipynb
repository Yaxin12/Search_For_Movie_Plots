set up model

```python
model_name = 'nq-distilbert-base-v1'
bi_encoder = SentenceTransformer(model_name)
top_k = 5
```

Encode the dataset

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
plot_embeddings = bi_encoder.encode(plots)
title_embeddings = bi_encoder.encode(titles)
```

```python
def search(query):
    query_embedding = bi_encoder.encode([query])[0]

    # Calculate cosine similarity between query and plot embeddings
    plot_similarities = cosine_similarity([query_embedding], plot_embeddings).flatten()
    plot_indices = plot_similarities.argsort()[::-1][:top_k]  # Top k most similar plots

    plot_results = [plots[i] for i in plot_indices]

    return plot_results
```

```python
query = "Western romance"
plot_results = search(query)
print("Results for query:", query)
for i, plot in enumerate(plot_results):
    print(f"{i+1}. {plot}")
```

Plots and Title

```python
def search(query):
    query_embedding = bi_encoder.encode([query])[0]
    title_embeddings = bi_encoder.encode(titles)

    # Calculate cosine similarity between query and plot embeddings
    plot_similarities = cosine_similarity([query_embedding], plot_embeddings).flatten()
    plot_indices = plot_similarities.argsort()[::-1][:5]  # Top 5 most similar plots

    # Calculate cosine similarity between query and title embeddings
    title_similarities = cosine_similarity([query_embedding], title_embeddings).flatten()
    title_indices = title_similarities.argsort()[::-1][:5]  # Top 5 most similar titles

    plot_results = [plots[i] for i in plot_indices]
    title_results = [titles[i] for i in title_indices]

    return plot_results, title_results
```

```python
# All queries
queries = [
    """Documentaries showcasing indigenous peoples' survival and daily life in Arctic regions""",
    """Western romance""",
    """Silent film about a Parisian star moving to Egypt, leaving her husband for a baron, and later reconciling after finding her family in poverty in Cairo.""",
    """Comedy film, office disguises, boss's daughter, elopement.""",
    """Lost film, Cleopatra charms Caesar, plots world rule, treasures from mummy, revels with Antony, tragic end with serpent in Alexandria.""",
    "Denis Gage Deane-Tanner"
]

ground_truth = [
    'Nanook of the North',
    'The Lucky Horseshoe',
    'Sahara',
    'Ask Father',
    'Cleopatra',
    'Captain Alvarez'
]
prediction = []

for query in queries:
    # Call the search function
    plot_results, title_results = search(query)

    # Print the results
    print("Results for query:", query)
    # print("Top 5 most similar plots:")
    for i, plot in enumerate(plot_results):
        print(f"{i+1}. {plot}")
    print("\nTop 5 most similar titles:")
    for i, title in enumerate(title_results):
        print(f"{i+1}. {title}")

    prediction.append(title_results)
```

```python
# Recall@1
def recall(pred, gt, rank = 1):
  counter = 0
  for p, t in zip(pred, gt):
    if t in p[:rank]:
      counter += 1
  return counter / len(gt)

recall(prediction, ground_truth)
```

```python
#MMR
def mmr(pred, gt):
  sumrank = 0
  for p, t in zip(pred, gt):
    try:
      rank = 1/(p.index(t) + 1)
    except ValueError:
      rank = 0
    sumrank += rank
  return sumrank/len(gt)

mmr(prediction, ground_truth)
```
